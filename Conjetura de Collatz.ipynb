# Conjetura de Collatz

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
sns.set()
```

```python
def next_number(number):
    if number % 2 == 0:
        return number / 2
    
    else:
        return (number * 3) + 1
```

```python
def get_results(initial):
    results = []
    while initial != 1:
        initial = next_number(initial)
        results.append(initial)
        
    return results
```

```python
initial = 27
continuar = True
results = get_results(initial)
end = (len(results)-1)

plt.figure(figsize=(15,8))
plt.plot(results)
plt.scatter(end,1,color="r", marker=".", s=150)
plt.text(end, 300, "1,1", fontsize=15)
plt.title("3N + 1", fontsize=20)
plt.xlabel("Iteraciones", fontsize=18)
plt.ylabel("Numero", fontsize=18)
plt.show()
```

Si realizamos la prueba con distintos numeros podremos observar que cualquier numero que siga las reglas establecidas en algun momento terminan en 1

```python
fig, ax = plt.subplots(2,2, figsize=(15,8))
numbers = [27, 100, 987, 1573]

for i, ax_i in enumerate(fig.get_axes()):
    results = get_results(numbers[i])
    end = (len(results)-1)
    
    ax_i.plot(results)
    ax_i.scatter(end,1,color="r", marker=".", )
    ax_i.text(end, 1, "1,1", fontsize=10)
    ax_i.set_title("Numero {}, con {} iteraciones".format(numbers[i], len(results)), fontsize=15)
    
plt.show()
```
