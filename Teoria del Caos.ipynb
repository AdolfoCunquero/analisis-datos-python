# Teoria del Caos

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
sns.set()
```

```python
def calc_x(r, x):
    return r * x * (1 - x)
```

```python
def calc_r(r, years):
    result = []
    initial = 0.4

    for i in range(years):
        initial = calc_x(r, initial)
        result.append(initial)
        
    return result
```

Cuando r <= 1, la tasa de crecimiento es muy baja y la poblacion tiende a extinguirse, pero cuando es mayor a 1 con el paso de los aÃ±os tiende a estabilizarse

```python
fig = plt.figure(figsize=(15,8))
r = np.arange(0,3.0,0.5)
years = 101

for i in r:
    results = calc_r(i, years)
    plt.plot(results, label="r = {}".format(i))
    
plt.legend(loc="upper right", fontsize=15)
plt.title("Tasa de crecimiento a lo largo del tiempo", fontsize=18)
plt.xlabel("Tiempo transcurrido", fontsize=15)
plt.ylabel("Tasa de crecimiento", fontsize=15)
plt.show()
```

Cuando r > 3.0 no se estabiliza en un solo valor, mientras mas grande sea r, habran mas puntos de estabilidad antes de comenzar de nuevo el ciclo

```python
years = 21
fig, ax = plt.subplots(3,2, figsize=(15,8), sharex=True, sharey=True)
plt.figure(figsize=(18,9))

r = np.linspace(3.4, 3.9, 6)

for i, ax_i in enumerate(fig.get_axes()):
    results = calc_r(r[i], years)
    ax_i.plot(results, marker="o", label="r = {}".format(r[i]))
    ax_i.set_title("r = {}".format(round(r[i],2)), fontsize=15)
    
plt.show()
```

Si colocamos r en el eje x y el crecimiento en el eje y, agrupando los datos podremos ver el comportamiento de la estabilidad de la poblacion.

```python
r = np.arange(0.0,1,0.1)
years = 101
results = []
for i in r:
    results.append(calc_r(i, years))

plt.plot(r, results, 'g.')
plt.show()
```

Grafico completo

```python
r = np.arange(0.0,4.0,0.01)
years = 101
results = []
for i in r:
    results.append(calc_r(i, years))
    
plt.figure(figsize=(15,8))
plt.plot(r,results, 'b')
plt.ylim(0,1)
plt.xlabel("Tasa de crecimiento", fontsize=18)
plt.ylabel("Poblacion", fontsize=18)
plt.title("Crecimiento de la poblacion respecto a r", fontsize=20)
plt.show()
```

Si nos enfocamos al comportamiento que tiene el crecimiento de la poblacion vemos que cuando r o la tasa de creciminto es menor a 1 tiende a serguir en cero, cuando supera ese valor, la poblacion se estabiliza en un valor mas grande, al superar 3 se comienza a dividir en dos ramas continuamente, hasta alcanzar el caos total despues de 3.6

```python
r = np.arange(0.0,4.0,0.01)
years = 101
results = []
for i in r:
    results.append(calc_r(i, years))
    
plt.figure(figsize=(15,8))
plt.plot(r,results, 'b', alpha=0.05, lw=1)
plt.ylim(0,1)
plt.xlabel("Tasa de crecimiento", fontsize=18)
plt.ylabel("Poblacion", fontsize=18)
plt.title("Teoria del caos", fontsize=20)
plt.show()
```
