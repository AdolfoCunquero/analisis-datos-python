# Triangulo de Sierpinski

```python
import matplotlib.pyplot as plt
import random
```

```python
x = [0,10,5]
y = [0,0,8]
```

```python
def get_next_vertice():
    return random.randint(0,2)
```

```python
x_seed = 4
y_seed = 2

x.append(x_seed)
y.append(y_seed)

for i in range(100000):
    vertice = get_next_vertice()
    x_seed = (x[vertice] + x_seed)/2
    y_seed = (y[vertice] + y_seed)/2
    x.append(x_seed)
    y.append(y_seed)
    
```

```python
plt.figure(figsize=(9,9))
plt.scatter(x,y,s=1)
plt.title("Triangulo de Sierpinski", fontsize=25)
plt.show()
```
